```python
import numpy as np
import matplotlib.pyplot as plt


ax = plt.figure(figsize=(16,9)).add_subplot(projection='3d')
data = np.loadtxt("datos.txt", skiprows=1)

x = data[:,1]
y = data[:,2]
z = data[:,3]


ax.plot(x, y, z, label='parametric curve')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
