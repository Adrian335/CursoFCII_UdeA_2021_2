```python
import pandas as pd
import matplotlib.pylab as plt
import mpl_toolkits.mplot3d.axes3d as p3
```

```python
df = pd.read_csv ("figura.csv")
print(df)
```

```python
x=df["x"]
y=df["y"]
z=df["z"]
```

```python
plt.plot(x,y,"-o")
```

```python
# Creamos la figura
fig = plt.figure(figsize=(10,10))
# Creamos el plano 3D
ax1 = fig.add_subplot(111, projection='3d')
ax1.scatter(x, y, z)
plt.show()
```
